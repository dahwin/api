```python
from fastapi import FastAPI
from starlette.responses import StreamingResponse
import cv2
import threading
from queue import Queue
import asyncio
import nest_asyncio

nest_asyncio.apply()
app = FastAPI()

# Queue to store and distribute the latest frame
frame_queue = Queue(maxsize=1)

def generate_frames():
    url = "http://localhost:3000/video_feed"  # Update the URL if needed
    cap = cv2.VideoCapture(url)
    while True:
        ret, frame = cap.read()
        if ret:
            _, buffer = cv2.imencode('.jpg', frame)
            latest_frame = buffer.tobytes()
            if frame_queue.full():
                frame_queue.get_nowait()  # Remove the old frame if the queue is full
            frame_queue.put(latest_frame)  # Add the latest frame to the queue
        else:
            frame_queue.put(None)  # If frame reading fails, set frame_queue to None

def start_frame_generation():
    # Start the frame generation in a separate thread
    threading.Thread(target=generate_frames, daemon=True).start()

@app.get("/video_feed")
async def video_feed():
    async def stream_frame():
        while True:
            frame = frame_queue.get()  # Retrieve the latest frame from the queue
            if frame:
                yield (
                    b'--frame\r\n'
                    b'Content-Type: image/jpeg\r\n\r\n' + frame + b'\r\n'
                )
            else:
                await asyncio.sleep(0.02)  # If no frame available, wait for a short time

    return StreamingResponse(stream_frame(), media_type="multipart/x-mixed-replace; boundary=frame")

if __name__ == "__main__":
    start_frame_generation()  # Start frame generation in a separate thread
    import uvicorn
    uvicorn.run(app, host="127.0.0.1", port=8000)
```

```python
from fastapi import FastAPI
from starlette.responses import StreamingResponse
import cv2
import threading
from queue import Queue
import asyncio
import nest_asyncio
import socket

nest_asyncio.apply()
app = FastAPI()

# Queue to store and distribute the latest frame
frame_queue = Queue(maxsize=2)

def generate_frames():
    url = "http://localhost:3000/video_feed"  # Update the URL if needed
    cap = cv2.VideoCapture(url)
    while True:
        ret, frame = cap.read()
        if ret:
            _, buffer = cv2.imencode('.jpg', frame)
            latest_frame = buffer.tobytes()
            if frame_queue.full():
                frame_queue.get_nowait()  # Remove the old frame if the queue is full
            frame_queue.put(latest_frame)  # Add the latest frame to the queue
        else:
            frame_queue.put(None)  # If frame reading fails, set frame_queue to None

def start_frame_generation():
    # Start the frame generation in a separate thread
    threading.Thread(target=generate_frames, daemon=True).start()

@app.get("/video_feed")
async def video_feed():
    async def stream_frame():
        while True:
            try:
                frame = frame_queue.get()
                if frame:
                    yield (
                        b'--frame\r\n'
                        b'Content-Type: image/jpeg\r\n\r\n' + frame + b'\r\n'
                    )
            except Exception as e:
                # Handle exceptions (log, reconnect, etc.)
                print(f"Error sending frame: {e}")
                break  # Exit the loop on critical errors
            else:
                await asyncio.sleep(0.02)  # Wait for a short time

    return StreamingResponse(stream_frame(), media_type="multipart/x-mixed-replace; boundary=frame")

if __name__ == "__main__":
    start_frame_generation()  # Start frame generation in a separate thread
    import uvicorn
    uvicorn.run(app, host="127.0.0.1", port=8000)
```
